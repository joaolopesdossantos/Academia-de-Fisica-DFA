## Links e Imagens

### Links

A forma de um link em Markdown é 

`[texto](pasta/nome_de_ficheiro)` 

por exemplo para ver um ficheiro que está na pasta pdfs

[ficheiro](pdfs/Intro_JupyterNotebook-estudante.pdf)

se te  aparecer o erro `404 : NOT FOUND` deves ter um erro pasta ou nome do ficheiro.

Quando descarregas o notebook do Github pode aparecer um caminho alterado

`[ficheiro](files/pdfs/Intro_JupyterNotebook-estudante.pdf)`

No meu sistema (versão do Jupyter 6.4.12) esta forma do link funciona. Se no teu não funcionar, experimenta retirar `files` anstes do nome da pasta. 

Este mesmo esquema funciona para obter um ficheiro da web:

[notas de mecânica](https://www.fc.up.pt/pessoas/jlsantos/wp-content/myuploads/FermiProblems.pdf)

ou ligar um site externo

[PORDATA](https://www.pordata.pt/)

Em `Markdown`   os URL ou nomes de ficheiros **não** são colocados entre plicas. 


###  Imagens

Uma figura vale mil palavras, por isso deve haver maneira de inserir figuras num notebook. Na verdade há várias.

##### Arrastamento de ficheiro de imagem

Na versão que uso do jupyter notebook, 6.4.12, existe no menu `Edit` a opção `Insert Image`. Nas versões mais recentes esta opção desapareceu. Mas podes inserir uma imagem **numa célula markdown** arrantando o ficheiro  (`.png`,`.gif`,`.jpg`) para a mesma. 

A imagem seguinte foi inserida desse modo



O texto que aparece é ``. O código da imagem é inserido na própria página web. Mesmo que o ficheiro da imagem seja apagado, a imagem permanece no notebook.  A cópia deste texto *para outra célula* não  carrega a imagem (experimenta).

#####  Inserção do link em Markdown
Podes sempre inserir um link para uma imagem como fizemos acima para outros ficheiros. Se este for precedido `!`  a imagem é vizualizada no notebook.  Sem este carater verás apenas o link para a figura que a abre noutra janela.

#### HTML

Em alternativa podes usar código `html`. Faz duplo clique na célula para veres o código.
<img src="Images/porto-2537851_1280.jpg" width=300 align='center'/>

Em `html` podes controlar o tamanho e a posição da imagem. 

**Notas:** 
 - Em `html` a pasta e nome do ficheiro, ou o enderreço web, devem estar entre `"..."`; em `Markdown` não.
 - Evita caracteres especiais (hifen, - , underscore,\_, ou espaços) no nome dos ficheiros de figuras.
