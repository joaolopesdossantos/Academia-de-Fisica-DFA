### Gráficos
Antes de deixarmos esta breve introdução vamos dar dois exemplos de criação de gráficos. Mais tarde vampos perceber o código. Para já limita-te a tentar antecipar o resultado da execução de cada célula. 

```python
%matplotlib inline                         
import numpy as np                         # modulo numérico , essencial para trabalho cientifico; chamado
                                           # com nome abreviado np
from matplotlib.pylab import *             # modulo de graficos. 
x=[1,2,3,4,5]                              # abcissas
y=[1,1,3,2,2]                              # ordenadas
plot(x,y);
```

```python
# gráfico de 2 funções
x= np.linspace(0,10,100)               # lista de abcissas de 0 a 10 com 100 pontos
y = np.sin(x)                          # lista de ordenadas; a função sin é aplicada a cada elemento da lista x
z= np.cos(x)                           # lista de ordenadas; a função cos é aplicada a cada elemento da lista x
plot(x,y,'bo',x,z,'g^')  
```

```python
x= np.linspace(0,10,11) 
print(x)
```

```python
# Um gráfico mais arranjadinho
figure(figsize=(4,2))
xlabel("x", fontsize=12)
ylabel ("sin(x)", fontsize=12)
title("Gráfico de sin(x)", fontsize=14)
grid(True)
plot(x,y);
```

### Escrita de equações
Se usares o Jupyter para Física ou matemática, mais tarde ou mais cedo, provavelmente mais cedo que mais tarde, terás de escrever equações. O Jupyter suporta o sistema LaTeX usado por físicos, matemáticos e cientistas de computadores, e por toda a gente que se preocupa em escrever equações bem legíveis. Em LaTeX indica-se  a entrada em modo matemático envolvendo o código da equação entre   $\$ $. e $\$$. Os comandos LaTeX  começam sempre com `\`.
Eis alguns exemplos

$\alpha,\beta,\gamma$

$\sqrt{1-x^2}$

$ P(x)=a_0 +a_1 x+a_2 x^2$

Se em vez de um $\$$  inserires o código entre pares $\$\$$ e $\$\$$ , eis o que acontece
$$ \vec{F}=\vec{a}$$

$$
\frac{1}{\sqrt{1-x}}
$$

Diverte-te a inventar mais equações. Podes encontrar uma lista de comandos LaTeX em
